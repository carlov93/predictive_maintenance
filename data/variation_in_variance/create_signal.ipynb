```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import signal
from math import exp
from math import pi
```

## Parameter

## Create Signal with variation in variance

```python
x = np.random.uniform(-0.1,1, 1000)
```

```python
y = []
mu = []
sigma = [] 
```

```python
time = np.linspace(-3,1,3000)
for index, i in enumerate(time):
    mu_i = 2 * exp(-1*i)*np.cos(4*pi*i)
    sigma_i = min(2*index*0.001,5)
    y_i = np.random.normal(mu_i, sigma_i)
    y.append(y_i)
    mu.append(mu_i)
    sigma.append(sigma_i)
```

```python
y = np.asarray(y)
mu = np.asarray(mu)
sigma = np.asarray(sigma)
```

```python
fig = plt.figure(figsize=(15,6), dpi=200)
ax = fig.add_subplot(111)
sns.lineplot(x=time, 
             y=mu, 
             ax=ax, 
             color="black",
             label="$\hat{y}$",
             linewidth=.8)

lower_bound = mu - sigma
upper_bound = mu + sigma
ax.fill_between(time, lower_bound, upper_bound, color="grey", alpha=0.5, label="True Interval of $\sigma$")

sns.scatterplot(x=time, 
                y=y,
                ax=ax,
                color="blue",
                s=10,
                linewidth=0.1)


plt.show()
```

```python
df = pd.DataFrame(data={"ID": range(len(time)), "time":time, "mu":mu, "sigma":sigma, "y_true":y})
df.head()
```

## Save to csv file

```python
df.to_csv("./dataset_increasing_variance.csv", index=False, sep=",")
```
