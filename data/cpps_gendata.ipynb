## Requirements

*Requirement 1 (from (i)):* The number of latent resp.\ observable variables $m$ resp.\ $n$ should be flexible. Especially the dimension of the observation space $n$ will be very high. 

*Requirement 2 (from (i)):* Usually, there is a functional relation between the latent variables $\mathbf{x}(t)$, e.g. the power consumption of a drive is closely related to the drive's acceleration. One says that the latent variables connected via a characteristic curve. Often $f(\mathbf{x}(t)) = 0$ for a low degree polynomial $f$. Here different complexities of $f$ must be definable. 

*Requirement 3 (from (ii),  (iii)):* The data should be generated as time series data where consecutive data points are usually in close proximity of each other, representing the continuous behavior of the plant. Optionally, state variables may lead to a long time trend in the time series.

*Requirement 4 (from (ii):* At the occurrence of certain events, the data generation process should be reset and changed, modeling discontinuous mode changes of the system.

## Parameters

```python
export = True # Export data as csv in wd
export_img = True
separator = ';'

dim_latent_space = 3 # Number of dimensions in latent space (don't set anythign else than 3)
num_samples = 50000 # Number of samples in latent space
thres_latent_space = 2 * np.pi # Threshold value in latentspace for all dimensions
degree_cc = 3 # Characteristic curve is polynomial of degree degree_cc  (don't set anythign else than 3)
max_cc_param = 1.0 # Maximum curviness of CC
random_walk_type = 'const' # Type of random walk. Choose 'const' or 'rand' distance
max_walk_distance_param = thres_latent_space / 100 # Max walk distance
walk_bias = 100.0 # bias towards direction of last step

# Where to start with adding errors (start, end)
error_latent_space_small_pos = np.random.randint(int(np.floor(num_samples * 0.4)),
                                                 int(np.floor(num_samples * 0.5)))
error_latent_space_large_pos = np.random.randint(int(np.floor(num_samples * 0.7)),
                                                 int(np.floor(num_samples * 0.8)))

# Error gradient in latent space per dimension (mean, spread)
error_latent_space_small = [max_walk_distance_param * 0.000009, max_walk_distance_param * 0.000009]
error_latent_space_large = [max_walk_distance_param * 0.00003, max_walk_distance_param * 0.00001]
errors = []

dim_observation_space = 10 # Number of dimensions in observation space

# Type of mapping function from latent to observation space.
# Choose 'linear', 'sinusoid', 'exponential' or 'all'
type_mapping = 'sinusoid'

max_obs_param = 5.0 # Threshold for mapping parameters to observation space

max_value = 1
```

```python
#hide_me

def export_csv(data, num_columns, filename):
    file = open(filename, 'w+')
    
    for i in range(int(len(data) / num_columns)):
        input = ''
        for j in range(num_columns):
            input += str(data[i * num_columns + j])
            if j < num_columns - 1:
                input += separator
        input += '\n'
        file.write(input)
    
    file.close()
    
    print(''.join(['Export to ', filename]))
```

## Characteristic curve

First, in the m-dimensional latent space, a characteristic curve $f(\mathbf{x})=0$ is generated as a polynomial of degree 3, the parameters of the polynomial are chosen randomly. The data points on this zero set represent the normal system behavior.

```python
def gen_cc_param():
    param = np.random.uniform(-max_cc_param,max_cc_param,dim_latent_space * (degree_cc + 1))
    
    # f(x) = 0
    for i in range(degree_cc, len(param), (degree_cc + 1)):
        param[i] = 0.0
            
    return param

cc_param = gen_cc_param()
```

```python
#hide_me

if export:
    filename = ''.join(['./data/parameters', '.txt'])
    file = open(filename, 'a')

    file.write(''.join(['cc_param = ', str(cc_param), '\n']))

    file.close()
```

```python
@jit(nopython=True)
def get_x_last(x_array_dim_minus_one):
    x_last = 0.0

    for i in range(dim_latent_space - 1):
        for j in range((degree_cc + 1)):
            index_param = (dim_latent_space+1) * i + j
            x_last += cc_param[index_param] * np.power(x_array_dim_minus_one[i], degree_cc - j)

    x_last /= np.power(thres_latent_space, degree_cc)
    return x_last

def cc(x_array_dim_minus_one):
    x_array = np.append(x_array_dim_minus_one, get_x_last(x_array_dim_minus_one))
    
    return x_array
```

```python
#hide_me

@jit(nopython=True)
def cc3d(x,y):
    z = cc_param[0] * np.power(x, 3) + cc_param[1] * np.power(x, 2) + cc_param[2] * x + cc_param[3] + \
        cc_param[4] * np.power(y, 3) + cc_param[5] * np.power(y, 2) + cc_param[6] * y + cc_param[7]
    z /= np.power(thres_latent_space, degree_cc)
    return z

def get_data_cc3d():
    samples = 50
    x = np.linspace(-thres_latent_space, thres_latent_space, num=samples)
    y = np.linspace(-thres_latent_space, thres_latent_space, num=samples)
    X, Y = np.meshgrid(x, y)
    
    Z = cc3d(X, Y)
    
    return X,Y,Z

def plot_cc3d():
    X,Y,Z = get_data_cc3d()
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
    ax.set_title('Characteristic curve');

    plt.show()

if dim_latent_space == 3:
    plot_cc3d()
```

## Random walk on CC

To fulfill requirement 4, when the data points in latent space move beyond a predefined threshold, either in time or in value, the data generation starts again at a random point on the characteristic curve.

```python
def get_starting_point_cc():
    starting_point = np.random.uniform(-thres_latent_space,thres_latent_space, dim_latent_space - 1)
    
    return cc(starting_point)
```

```python
def next_random_walk_cc(prev_point, last_walk):
    # generate normalized vector
    new_random_walk = np.random.uniform(-1, 1, dim_latent_space - 1)
    new_random_walk /= np.linalg.norm(new_random_walk)
    
    if random_walk_type == 'const':
        new_random_walk *= max_walk_distance_param
    else:
        new_random_walk *= np.random.rand(max_walk_distance_param)
    
    new_walk = (walk_bias * last_walk + new_random_walk) / (walk_bias + 1.0 )
    
    # check if this is the first step
    if np.sum(last_walk) == 0.0:
        new_walk = new_random_walk
    
    last_walk = new_walk
    next_point = prev_point[:dim_latent_space - 1] + new_walk
    return cc(next_point)
```

```python
def gen_latent_space(space, last_walk):
    for i in range(num_samples - 1):
        new_sample = next_random_walk_cc(space[-dim_latent_space:], last_walk)

        # check threshold
        for i in range(len(new_sample) - 1):
            if new_sample[i] >= thres_latent_space or new_sample[i] <= -thres_latent_space:
                new_sample = get_starting_point_cc()
                break

        space = np.append(space, new_sample)
    return space
```

```python
samples_latent_space = get_starting_point_cc()
samples_latent_space_train = get_starting_point_cc()

last_walk_test = np.zeros(dim_latent_space - 1)
last_walk_train = np.zeros(dim_latent_space - 1)

samples_latent_space = gen_latent_space(samples_latent_space, last_walk_test)
samples_latent_space_train = gen_latent_space(samples_latent_space_train, last_walk_train)
```

```python
#hide_me

def plot3d(data, title):
    if dim_latent_space == 3:
        x,y,z = np.reshape(data, (dim_latent_space, num_samples), order='F')
        X,Y,Z = get_data_cc3d()

        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none', alpha=0.5)
        ax.scatter(x, y, z, color='red', alpha=1.0, s=0.01)
        ax.set_title(title);

        if export_img:
            filename = ''.join(['./data/', title.replace(' ', '_').replace('(', '').replace(')', ''), '_','.pdf'])
            plt.savefig(filename.lower())
        else:
            plt.show()
```

## Generate error set

```python
errors = np.zeros(dim_latent_space * num_samples)

for i in range(num_samples):
    error = np.zeros(dim_latent_space)
    
    if i > error_latent_space_large_pos:
        error = np.random.normal(error_latent_space_large[0] * i,
                                 error_latent_space_large[1] * i,
                                 dim_latent_space)
    elif i > error_latent_space_small_pos and i <= error_latent_space_large_pos:
        error = np.random.normal(error_latent_space_small[0] * i,
                                 error_latent_space_small[1] * i,
                                 dim_latent_space)
    
    errors[i*dim_latent_space:(i+1)*dim_latent_space] = error
    
samples_latent_space_error = samples_latent_space + errors
```

```python
plot3d(samples_latent_space_error, 'Random walk on CC with errors')
```

## Map to observation space

By means of a function $y$, $X$ is mapped from an $m$-dimensional latent space into an $n$-dimensional observation space ($m \ll n$). A multi-valued function $g \in \{g_1, g_2, g_3\}^n$ is chosen, where $g_1$ is a linear function with random parameters, $g_2$ comprises linear and sinusoid terms with random parameters, $g_3$ adds exponential terms to $g_2$. This generates a data set $Y=g(X)$ via $\mathbf{y}_i=g(\mathbf{x}_i)\in\mathbb{R}^n$ of *ordinary* data points and fulfills requirement 1. 

```python
def gen_obs_param():
    obs_param = np.random.uniform(-max_obs_param, max_obs_param, dim_latent_space * 3 * dim_observation_space)
    
    return obs_param

obs_param = gen_obs_param()
```

```python
#hide_me

if export:
    filename = ''.join(['data/parameters','.txt'])
    file = open(filename, 'a')

    file.write(''.join(['obs_param = ', str(obs_param), '\n']))

    file.close()
```

```python
@cuda.jit
def gen_obs_space_sinusoid(samples_obs_space, samples_latent_space,
                           dim_latent_space, dim_observation_space, obs_param,
                           max_obs_param, max_value):

    pos = cuda.grid(1)
    if pos < samples_obs_space.size:
        latent_index = int(
            math.floor(pos / dim_observation_space) * dim_latent_space)
        x = samples_latent_space[latent_index]
        y = samples_latent_space[latent_index + 1]
        z = samples_latent_space[latent_index + 2]

        obs_index = pos % dim_observation_space

        para_abc = 1.0 / max_obs_param * max_value / 3.0 / thres_latent_space
        a = obs_param[0 * obs_index] * para_abc
        b = obs_param[1 * obs_index] * para_abc
        c = obs_param[2 * obs_index] * para_abc

        para_def = 1.0 / max_obs_param * max_value / 3.0
        d = obs_param[3 * obs_index + dim_latent_space] * para_def
        e = obs_param[4 * obs_index + dim_latent_space] * para_def
        f = obs_param[5 * obs_index + dim_latent_space] * para_def

        para_ghi = 1.0 / max_obs_param * max_value / 3.0 / math.pow(
            thres_latent_space, 2)
        g = obs_param[3 * obs_index + 2 * dim_latent_space] * para_ghi
        h = obs_param[4 * obs_index + 2 * dim_latent_space] * para_ghi
        i = obs_param[5 * obs_index + 2 * dim_latent_space] * para_ghi

        samples_obs_space[pos] += a * x + b * y + c * z + d * math.sin(
            10.0 * x) + e * math.sin(10.0 * y) + f * math.sin(
                10.0 * z) + g * x * y + h * y * z + i * z * x
```

```python
size_obs_space = (int)(len(samples_latent_space) / dim_latent_space * dim_observation_space)

threadsperblock = 256
blockspergrid = math.ceil(size_obs_space / threadsperblock)
```

```python
if type_mapping == 'all' or type_mapping == 'sinusoid':
    samples_obs_space_sin = np.zeros(size_obs_space)
    samples_obs_space_train_sin = np.zeros(size_obs_space)
    samples_obs_space_error_sin = np.zeros(size_obs_space)
    
    gen_obs_space_sinusoid[blockspergrid, threadsperblock](
        samples_obs_space_sin, samples_latent_space, dim_latent_space,
        dim_observation_space, obs_param, max_obs_param, max_value)
    
    gen_obs_space_sinusoid[blockspergrid, threadsperblock](
        samples_obs_space_train_sin, samples_latent_space_train, dim_latent_space,
        dim_observation_space, obs_param, max_obs_param, max_value)
    
    gen_obs_space_sinusoid[blockspergrid, threadsperblock](
        samples_obs_space_error_sin, samples_latent_space_error, dim_latent_space,
        dim_observation_space, obs_param, max_obs_param, max_value)
```
