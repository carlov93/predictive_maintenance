```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
history = pd.read_csv("./loss_MLE_2.csv")
```

```python
history_1 = history.iloc[:5000,:]
history_2 = history.iloc[5000:,:]
```

```python
f, ax = plt.subplots(figsize=(15,7.5), dpi=200)
sns.lineplot(data=history_1, x=range(history_1.shape[0]), y="loss iteration", ax=ax, label="training")
ax.set(xlabel='Training Iteration', ylabel='loss')
plt.savefig("./loss_fluctuation")
```

```python
f, ax = plt.subplots(figsize=(15,7.5), dpi=200)
sns.lineplot(data=history_2, x=history_2.iloc[:,0], y="loss iteration", ax=ax, label="training")
plt.ylim(0,500)
plt.text(5380,470, 'max_loss = 40107', fontsize=12)
ax.set(xlabel='Training Iteration', ylabel='loss')
plt.savefig("./loss_instability")
```

## Analyse Sigma / exp(tau)

```python
history = pd.read_csv("./loss_MLE_tau.csv")
```

```python
history_1 = history.iloc[:5000,:]
history_2 = history.iloc[5000:,:]
```

```python
f, ax = plt.subplots(figsize=(15,7.5), dpi=200)
sns.lineplot(data=history_1, x=range(history_1.shape[0]), y="tau", ax=ax, label="training")
plt.ylim(0, 5)
plt.text(3000,4.5, 'max_sigma = 25', fontsize=12)
ax.set(xlabel='Training Iteration', ylabel='sigma = exp(tau)')
plt.savefig("./loss_tau_1")
plt.plot()
```

```python
f, ax = plt.subplots(figsize=(15,7.5), dpi=200)
sns.lineplot(data=history_2, x=range(5000,history_2.shape[0]+5000), y="tau", ax=ax, label="training")
plt.ylim(0, 10)
plt.text(6200,9, 'max_sigma = unendlich', fontsize=12)
ax.set(xlabel='Training Iteration', ylabel='exp(tau)')
plt.savefig("./loss_tau_2")
plt.plot()
```
