```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from matplotlib.pyplot import figure
```

# Visualisation of forcast's drift
## Phase 1 - no error

```python
prediction_y = pd.read_csv("../../visualisation/files/prediction/MLE/cpps_degradation.csv", sep=";")
subset_y = prediction_y.iloc[:1000,:]
```

```python
prediction_yhat = pd.read_csv("../../visualisation/files/prediction/MLE/cpps_degradation_yhat.csv", sep=";")
subset_yhat =prediction_yhat.iloc[:1000,:]
```

```python
fig, axes = plt.subplots(10, 1, figsize=(30,40))
ax = axes.ravel()
columns_yhat = prediction_yhat.columns
columns_y = prediction_y.columns
for i in range(1, 11):
    sns.lineplot(data=subset_yhat, 
                         x=prediction_yhat.iloc[:,0], 
                         y=columns_yhat[i+10], 
                         ax=ax[i-1], 
                         color="red")
    sns.lineplot(data=subset_y, 
                         x=prediction_y.iloc[:,0], 
                         y=columns_y[i+10], 
                         ax=ax[i-1], 
                         color="black")
    ax[i-1].set_xlim((200,1000))
plt.show()
```

```python
prediction_y = pd.read_csv("../../visualisation/files/prediction/MLE/cpps_degradation.csv", sep=";")
subset_y = prediction_y.iloc[100:1000,:]
```

```python
prediction_yhat = pd.read_csv("../../visualisation/files/prediction/MLE/cpps_degradation_yhat.csv", sep=";")
subset_yhat =prediction_yhat.iloc[:1000-100,:]
```

```python
for i in range(1, 11):
    y = subset_y.iloc[:,i].values
    yhat = subset_yhat.iloc[:,i].values
    difference = (y-yhat)**2
    mse = 1/(subset_yhat.shape[0]-100) * np.sum(difference)
    print("Sensor No. {}:".format(i))
    print("MSE: {}".format(mse))
    print("- -"*10)
```
