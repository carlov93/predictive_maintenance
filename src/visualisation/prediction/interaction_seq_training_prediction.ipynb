```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from matplotlib.pyplot import figure
```

```python
data_50_5 = pd.read_csv("../files/prediction/pred_new_artifical_50_5.csv", sep=";")
data_50_50 = pd.read_csv("../files/prediction/pred_new_artifical_50_50.csv", sep=";")
data_50_150 = pd.read_csv("../files/prediction/pred_new_artifical_50_150.csv", sep=";")
data_150_5 = pd.read_csv("../files/prediction/pred_new_artifical_150_5.csv", sep=";")
data_150_50 = pd.read_csv("../files/prediction/pred_new_artifical_150_50.csv", sep=";")
data_150_150 = pd.read_csv("../files/prediction/pred_new_artifical_150_150.csv", sep=";")
```

```python
def visualize_prediction(dataframe):
    fig, axes = plt.subplots(2, 4, figsize=(15,7.5), dpi=200)
    ax = axes.ravel()
    columns = dataframe.columns
    plt.subplots_adjust(hspace=0.2, wspace=0.4)
    for i in range(1,9):
        sns.lineplot(data=dataframe, x="timestamp", y=columns[i], ax=ax[i-1], label="target")
        sns.lineplot(data=dataframe, x="timestamp", y=columns[i+8], ax=ax[i-1], label="predicted")
        ax[i-1].axvline(1536, 0,2, color="r", linestyle ="--", linewidth=3, label="train-validation")
        #plt.text(1570, 2, 'validation data', fontsize=12)
        #plt.text(1300, 2, 'training data', fontsize=12)
        ax[i-1].set(xlabel='time', ylabel='value')
        ax[i-1].set(title=columns[i])
    plt.show()
```

# Visualisation of Prediction - New Artifical Data
## Sensor Value Prediction - Train_seq:50 - Pred_seq:5

```python
visualize_prediction(data_50_5)
```

## Sensor Value Prediction - Train_seq:50 - Pred_seq:50

```python
visualize_prediction(data_50_50)
```
