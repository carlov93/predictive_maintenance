```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from matplotlib.pyplot import figure
```

```python
data_new_blade = pd.read_csv("../files/prediction/prediction_new_artifical_data.csv", sep=";")
data_residuals_worn = pd.read_csv("../files/residuals/residual_artifical_data.csv", sep=";")
data_worn_blade = pd.read_csv("../files/prediction/prediction_worn_artifical_data.csv", sep=";")
```

```python
def visualize_prediction(dataframe):
    fig, axes = plt.subplots(2, 4, figsize=(15,7.5), dpi=200)
    ax = axes.ravel()
    columns = dataframe.columns
    plt.subplots_adjust(hspace=0.2, wspace=0.4)
    for i in range(1,9):
        sns.lineplot(data=dataframe, x="timestamp", y=columns[i], ax=ax[i-1], label="target")
        sns.lineplot(data=dataframe, x="timestamp", y=columns[i+8], ax=ax[i-1], label="predicted")
        ax[i-1].axvline(1536, 0,2, color="r", linestyle ="--", linewidth=3, label="train-validation")
        #plt.text(1570, 2, 'validation data', fontsize=12)
        #plt.text(1300, 2, 'training data', fontsize=12)
        ax[i-1].set(xlabel='time', ylabel='value')
        ax[i-1].set(title=columns[i])
    plt.show()
```

# Visualisation of Prediction - New Artifical Data
## Sensor Value Prediction

```python
visualize_prediction(data_new_blade)
```

### Sine Wave Prediction

```python
fig, axes = plt.subplots(1, 1, figsize=(15,10))
columns = data_new_blade.columns
sns.lineplot(data=data_new_blade, x="timestamp", y=columns[8], ax=axes, label="target")
sns.lineplot(data=data_new_blade, x="timestamp", y=columns[8+8], ax=axes, label="predicted")
plt.show()
```

### Prediction loss

```python
fig, axes = plt.subplots(1,1, figsize=(15,10))
sns.lineplot(data=data_new_blade, x="timestamp", y=columns[17], label="prediction loss")
plt.show()
```

# Visualisation of Prediction - Worn Blade
## Sensor Value Prediction

```python
fig, axes = plt.subplots(2,4, figsize=(15,10))
ax = axes.ravel()
columns = data_worn_blade.columns
plt.subplots_adjust(hspace=0.2, wspace=0.4)
for i in range(1,9):
    sns.lineplot(data=data_worn_blade, x="timestamp", y=columns[i], ax=ax[i-1], label="target")
    sns.lineplot(data=data_worn_blade, x="timestamp", y=columns[i+8], ax=ax[i-1], label="predicted")
    ax[i-1].set(xlabel='time', ylabel='value')
    ax[i-1].set(title=columns[i])
plt.show()
```
