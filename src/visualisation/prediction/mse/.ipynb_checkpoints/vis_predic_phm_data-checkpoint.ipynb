```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from matplotlib.pyplot import figure
```

```python
def visualize_prediction(dataframe, title, line=False):
    fig, axes = plt.subplots(2, 4, figsize=(15,7.5), dpi=200)
    ax = axes.ravel()
    columns = dataframe.columns
    plt.subplots_adjust(hspace=0.2, wspace=0.4)
    for i in range(1,9):
        sns.lineplot(data=dataframe, x="timestamp", y=columns[i], ax=ax[i-1], label="target")
        sns.lineplot(data=dataframe, x="timestamp", y=columns[i+8], ax=ax[i-1], label="predicted")
        if line:
            ax[i-1].axvline(1536, 0,2, color="r", linestyle ="--", linewidth=3, label="train-validation")
        #plt.text(1570, 2, 'validation data', fontsize=12)
        #plt.text(1300, 2, 'training data', fontsize=12)
        ax[i-1].set(xlabel='time', ylabel='value')
        ax[i-1].set(title=columns[i])
    fig.suptitle(title)
    plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
def smooth_data(df, rate):
    for column in df.columns:
        smoothed_data = []
        for i,value in enumerate(df.loc[:,column]):
            if i==0:
                smoothed_data.append(value)
            else:
                x = rate  * value + (1 - rate) * smoothed_data[-1]
                smoothed_data.append(x)
        df.loc[:,column]=smoothed_data
    return df
```

# Visualisation of Prediction - New Machine
## Sensor Value Prediction

```python
data_new_blade = pd.read_csv("../files/prediction/NewBlade_with_sine.csv", sep=";")
visualize_prediction(data_new_blade, title="Prediction of New Machine", line=True)
plt.show()
```

### Sine Wave Prediction

```python
fig, axes = plt.subplots(1, 1, figsize=(15,10))
columns = data_new_blade.columns
sns.lineplot(data=data_new_blade, x="timestamp", y=columns[8], ax=axes, label="target")
sns.lineplot(data=data_new_blade, x="timestamp", y=columns[8+8], ax=axes, label="predicted")
axes.set_title("Prediction and Original Sine Signal without Errors")
plt.show()
```

### Prediction loss

```python
fig, axes = plt.subplots(1,1, figsize=(15,10))
sns.lineplot(data=data_new_blade, x="timestamp", y=columns[17])
axes.set_title("Prediction Loss New Machine")
plt.show()
```

# Visualisation of Prediction - Old Machine - Multiple Errors 
## Sensor Value Prediction

```python
data_worn_blade = pd.read_csv("../files/prediction/NewBlade_multiple_errors.csv", sep=";")
visualize_prediction(data_worn_blade, title="Prediction of Old Machine with Multiple Errors", line=False)
plt.show()
```

## Prediction Sine Wave with Errors

```python
fig, axes = plt.subplots(1, 1, figsize=(15,10))
sns.lineplot(data=data_worn_blade, x="timestamp", y=columns[8], ax=axes, label="target")
sns.lineplot(data=data_worn_blade, x="timestamp", y=columns[8+8], ax=axes, label="predicted")
plt.axvspan(200, 250, alpha=0.2, color='red')
plt.axvspan(400, 450, alpha=0.2, color='red')
plt.axvspan(700, 800, alpha=0.2, color='red')
plt.axvspan(1100, 1250, alpha=0.2, color='red')
plt.axvspan(1450, 1600, alpha=0.2, color='red')
plt.axvspan(1650, 1750, alpha=0.2, color='red')
axes.set_title("Prediction and Original Sine Signal with Multiple Errors")
plt.show()
```

## Prediction Loss

```python
fig, axes = plt.subplots(1,1, figsize=(15,10))
sns.lineplot(data=data_worn_blade, x="timestamp", y=columns[17], label="prediction loss")
axes.set_title("Prediction Loss Defective Machine")
plt.show()
```

## Visualisation Residuals of Sine Wave Sensor Data

```python
data_residuals_worn = pd.read_csv("../files/residuals/NewBlade_multiple_errors.csv", sep=";")
fig, axes = plt.subplots(1, 1, figsize=(15,10))
sns.scatterplot(data=data_residuals_worn, x="timestamp", y=data_residuals_worn[data_residuals_worn.columns[8]]) #scatter_kws={"s": 10})
plt.axhline(0.4, 0,2, color="r", linestyle ="-", linewidth=1)
plt.axhline(-0.4, 0,2, color="r", linestyle ="-", linewidth=1)
plt.axvspan(200, 250, alpha=0.3, color='red')
plt.axvspan(400, 450, alpha=0.3, color='red')
plt.axvspan(700, 800, alpha=0.3, color='red')
plt.axvspan(1100, 1250, alpha=0.3, color='red')
plt.axvspan(1450, 1600, alpha=0.3, color='red')
plt.axvspan(1650, 1750, alpha=0.3, color='red')
axes.set_title("Residuals of Prediction of Sine Signal with Multiple Errors")
plt.show()
```

# Visualisation of Prediction - Old Machine - Degradation Frequency
## Sensor Value Prediction

```python
data_degradation = pd.read_csv("../files/prediction/NewBlade_degradation_freq.csv", sep=";")
visualize_prediction(data_degradation, title="Prediction of Old Machine with Degradation",line=False)
plt.show()
```

```python
data_smoothed_loss = smooth_data(data_degradation, 0.01)
columns = data_smoothed_loss.columns
fig, axes = plt.subplots(1,1, figsize=(9,7))
sns.lineplot(data=data_smoothed_loss, x="timestamp", y=columns[17])
plt.axvspan(1000, 2000, alpha=0.2, color='red')
plt.ylim((0, 4))
axes.set_title("Smoothed Prediction Loss")
plt.show()
```

```python
data_residuals_worn = pd.read_csv("../files/residuals/NewBlade_degradation_freq.csv", sep=";")
fig, axes = plt.subplots(1, 1, figsize=(15,10))
sns.scatterplot(data=data_residuals_worn, x="timestamp", y=data_residuals_worn[data_residuals_worn.columns[8]])
plt.axhline(0.5, 0,2, color="r", linestyle ="-", linewidth=1)
plt.axhline(-0.5, 0,2, color="r", linestyle ="-", linewidth=1)
plt.axvspan(1000, 2000, alpha=0.2, color='red')
axes.set_title("Residuals of Prediction of Sine Signal with Increasing Frequency")
plt.show()
```

```python
data_smoothed_loss_shares = smooth_data(data_degradation, 0.05)
fig, axes = plt.subplots(2, 4, figsize=(15,7.5), dpi=200, sharey=True)
ax = axes.ravel()
columns = data_smoothed_loss_shares.columns
for i in range(18,26):
    sns.lineplot(data=data_smoothed_loss_shares, x="timestamp", y=columns[i], ax=ax[i-18])
    ax[i-18].axvspan(1000, 2000, alpha=0.2, color='red')
fig.suptitle("Share of Loss for each Signal")
plt.show()
```
