```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Visualisation - New Blade

```python
path_original = "../../data/vega_shrinkwrapper_original/NewBlade001_new_column_names.csv"
path_standardized = "../../data/vega_shrinkwrapper_standardized/NewBlade001.csv"
```

```python
df_original = pd.read_csv(path_original)
df_standardized = pd.read_csv(path_standardized)
```

## Histogramm - orignial data

```python
fig, axes = plt.subplots(2, 4, figsize=(15,15))
ax = axes.ravel()
plt.subplots_adjust(hspace=0.2, wspace=0.4)
for i in range(8):
    sns.distplot(df_original.iloc[:,i], ax=ax[i])
plt.show()
```

## Scatterplot - original data

```python
fig, axes = plt.subplots(2, 4, figsize=(15,15))
ax = axes.ravel()
plt.subplots_adjust(hspace=0.2, wspace=0.4)
for i in range(0,8):
    sns.scatterplot(x=df_original.iloc[:,0] , y=df_original.iloc[:,i], ax=ax[i])
plt.show()
```

## Pairplot - original data

```python
g = sns.PairGrid(df_original)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```
