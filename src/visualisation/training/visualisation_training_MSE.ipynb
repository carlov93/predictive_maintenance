```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Visualisation of loss during training
### MSE Loss Function 

```python
history = pd.read_csv("../files/history_training/history_MSE_with_tanh.csv",sep=";")
f, ax = plt.subplots(figsize=(15,7.5), dpi=200)
sns.lineplot(data=history, x="epoch", y="training", ax=ax, label="training")
sns.lineplot(data=history, x="epoch", y="validation", ax=ax, label="validation")
ax.set(xlabel='epoch', ylabel='loss')
#plt.savefig("../graphs/training_MSE_loss.png")
plt.show()
```

### MSE Loss Function - with dropout

```python
history = pd.read_csv("../files/history_training/history_MSE_dropout.csv",sep=";")
f, ax = plt.subplots(figsize=(10, 10))
sns.lineplot(data=history, x="epoch", y="training", ax=ax, label="training")
sns.lineplot(data=history, x="epoch", y="validation", ax=ax, label="validation")
ax.set(xlabel='epoch', ylabel='loss')
plt.show()
```
