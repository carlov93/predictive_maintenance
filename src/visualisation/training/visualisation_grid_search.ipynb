```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
style.use("seaborn")
```

```python
data=pd.read_csv("../files/grid_search/grid_search_stepsize.csv", sep=",")
```

## Results of Grid Search

```python
f, ax = plt.subplots(figsize=(15,7.5), dpi=200)
sns.scatterplot(x="Time", y="Loss", data=data, hue="LSTM Hidden", style="LSTM Layer", size="Sequence Length",
                ax=ax, palette="Reds", legend="full", sizes=(150,250))
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()
plt.show()
```
