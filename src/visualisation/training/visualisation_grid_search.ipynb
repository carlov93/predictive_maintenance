```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
style.use("seaborn")
```

```python
data=pd.read_csv("../files/grid_search/statistics_grid_search.csv", sep=",")
```

## Results of Grid Search

```python
f, ax = plt.subplots(figsize=(15,7.5), dpi=200)
sns.scatterplot(x="Time", y="Loss", data=data, hue="LSTM Hidden", style="LSTM Layer", size="Sequence Length",
                ax=ax, legend="full", palette="YlOrRd", sizes=(100, 300), markers=["o", "^"])
plt.gca().invert_yaxis()
plt.show()
```
