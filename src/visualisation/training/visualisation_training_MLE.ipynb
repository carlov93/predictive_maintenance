```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def visualize_training(path, path_save, save_graph=False):
    history = pd.read_csv(path,sep=";")
    f, ax = plt.subplots(figsize=(15,7.5), dpi=200)
    sns.lineplot(data=history, x="epoch", y="training", ax=ax, label="training")
    sns.lineplot(data=history, x="epoch", y="validation", ax=ax, label="validation")
    ax.set(xlabel='epoch', ylabel='loss')
    if save_graph:
        plt.savefig(path_save)
    plt.show()
```

## Visualisation of loss during training
### MSE Loss Function - Original Data - seq:50

```python
visualize_training("../files/history_training/history_MSE_dropout.csv", "../graphs/training_MSE_loss.png")
```

### MSE Loss Function - Artifical Data - seq:50

```python
visualize_training("../files/history_training/history_MSE_artifical_signal.csv", "../graphs/training_MSE_loss.png")
```

### MSE Loss Function - Artifical Data - seq:150

```python
visualize_training("../files/history_training/history_MSE_artifical_seq150.csv", "../graphs/training_MSE_loss.png")
```
