```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
import seaborn as sns
```

```python
path_original = "../../data/vega_shrinkwrapper_original/NewBlade/NewBlade001.csv"
df_original = pd.read_csv(path_original)
df_original.head()
```

## Autocorrelation

```python
fig, axes = plt.subplots(2, 4, figsize=(15,15))
ax = axes.ravel()
plt.subplots_adjust(hspace=0.2, wspace=0.4)
for i in range(8):
    plot_acf(df_original.iloc[:,i], lags=20, ax=ax[i])
    ax[i].set_title(df_original.columns[i])
    ax[i].set_xlabel('time lag')

    
plt.show()
```
