```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Visualisation - Worn Blade

```python
path_original = "../../../data/vega_shrinkwrapper_original/WornBlade/WornBlade001.csv"
df_original = pd.read_csv(path_original)
```

## Some statistics

```python
df_original.describe()
```

## Histogramm - orignial data

```python
fig, axes = plt.subplots(2, 4, figsize=(15,7.5), dpi=200)
ax = axes.ravel()
plt.subplots_adjust(hspace=0.2, wspace=0.4)
for i in range(8):
    sns.distplot(df_original.iloc[:,i], ax=ax[i])
plt.savefig("../graphs/raw_sensor_data_worn_blade_hist.png")
plt.show()
```

## Scatterplot - original data

```python
fig, axes = plt.subplots(2, 4, figsize=(15,7.5), dpi=200)
ax = axes.ravel()
plt.subplots_adjust(hspace=0.2, wspace=0.4)
for i in range(0,8):
    sns.lineplot(x=df_original.iloc[:,0] , y=df_original.iloc[:,i], ax=ax[i])
plt.savefig("../graphs/raw_sensor_data_worn_blade.png")
plt.show()
```
